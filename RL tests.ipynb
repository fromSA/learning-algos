# Q: 
Does Temporal Difference algorthim converge the value of states? 

```python
import numpy as np

env = [(0,1,False,0), (1,2,False,0), (1,5,False,0)]

# Initialize value function to zero
V = [0,0,0]

VV = []

# Set the number of episodes
num_episodes = 1

# Set the discount factor
gamma = 0.95

# Learning rate
alpha = 0.8

state =0

#for i_episode in range(num_episodes):
    # Reset the environment and get the first state
    #state = env.reset()

for t in range(200):
    action = t % 3

    # Perform the action and get the reward and next state
    next_state, reward, done, info = env[action]

    # TD Update
    td_target = gamma * (reward + V[next_state])
    td_error = td_target - V[state]
    V[state] = V[state] + alpha * td_error

    # Update the current state
    state = next_state

    VV.append(V.copy())

    # If the episode is done (we've reached a terminal state), then exit the loop for this episode
    #if done:
     #   break
        
# Print the estimated value function
print(V)
```
