```python
import stan
import numpy as np
import nest_asyncio
import asyncio

nest_asyncio.apply()
```

```python
LR = """
data{
    int<lower=0> N;
    real X[N];
}
parameters{
    real<lower=0> mean;
    real<lower=0> sigma;
}
model{
    target += normal_lpdf(mean| 0,1);
    target += normal_lpdf(sigma| 0,1);
    target += normal_lpdf(mean, sigma);
}
"""
```

```python
N =  100
X = np.random.normal(0, 1, N)

LR_data = {"N": N,"X": X}
```

```python
nest_asyncio.apply()
asyncio.run(asyncio.sleep(1))
```

```python
posterior = stan.build(LR, data=LR_data, random_seed=1)
```

```python
fit = posterior.sample(num_chains=4, num_samples=1000)
```

```python
eta = fit["eta"]
```

```python
df = fit.to_frame()
```
